```python
import json

in_path = "train.jsonl"          # 원본 파일
out_path = "train_stripped.jsonl"    # 결과 파일

with open(in_path, "r", encoding="utf-8") as fin, \
     open(out_path, "w", encoding="utf-8") as fout:
    
    for line in fin:
        line = line.strip()
        if not line:
            continue
        
        record = json.loads(line)
        
        instr = record.get("instruction", "")
        inp   = record.get("input", "")
        out   = record.get("output", "")
        
        # 1) Question: '?' 이후의 선택지(MoveA/MoveB...) 제거
        q = inp
        q_idx = q.rfind("?")
        if q_idx != -1:
            q = q[:q_idx + 1].strip()   # '?'까지 포함해서 자르고 공백 제거
        
        # 2) Answer: 'MoveB:b2b1' -> 'b2b1'만 남기기
        ans = out.strip()
        if ":" in ans:
            ans = ans.split(":", 1)[1].strip()
        
        new_record = {
            "instruction": instr,
            "question": q,
            "answer": ans
        }
        
        fout.write(json.dumps(new_record, ensure_ascii=False) + "\n")

print("완료! ->", out_path)
```
